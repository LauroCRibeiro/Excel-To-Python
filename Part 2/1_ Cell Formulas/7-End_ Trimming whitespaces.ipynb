# Trimming Whitespaces

## **Excel:**

    =TRIM(text)
<br>

## **Python:**

One string

    text.strip()
    text.lstrip()
    text.rstrip()

Series

    df['column'].str.strip()
    df['column'].str.lstrip()
    df['column'].str.rstrip()
<br>

### Load required packages and data

```
# Import required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
# Save Github location paths to a variable
failed_bank_path = 'https://github.com/The-Calculated-Life/python_analysis_for_excel/blob/main/data/failed_banks.xlsx?raw=true'

# Read excel and CSV files
bank_list = pd.read_excel(failed_bank_path, sheet_name='banks')

# Create messy_series
messy_series = pd.Series(['  Sometimes strings have whitespaces on both ends '
                          , ' Sometimes strings have leading whitespaces'
                          , 'Sometimes strings have trailing whitespaces'])
```

<br><br>

### Remove whitespace from both sides
---

```
# Remove whitespaces from " I am a string "
" I am a string ".strip()
```

<br>

```
# Use strip on bank names in bank_list
bank_list['Bank Name'].str.strip()
```

<br>

### Remove whitespace from one side only
---

```
# Remove left whitespace from " I am a string"
" I am a string".lstrip()
```

```
# Remove right whitespace from "I am a string "
"I am a string ".rstrip()
```
