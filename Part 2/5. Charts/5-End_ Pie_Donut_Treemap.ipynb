```
# Import required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import squarify
```

```
# Save Github location paths to a variable
bx_ratings_path = 'https://raw.githubusercontent.com/The-Calculated-Life/python_analysis_for_excel/main/data/bx_ratings.csv'

# Read excel and CSV files
bx_ratings = pd.read_csv(bx_ratings_path)

# Load data for chart
data = {'Category':['A', 'B', 'C', 'D', 'E', 'F', 'G'],
        'Value':[12, 8, 10, 6, 7, 8, 10]}

chart_data = pd.DataFrame(data)

rating_count = pd.DataFrame(bx_ratings.groupby('rating')['isbn'].count()).reset_index()
rating_count.columns = ['rating', 'book_count']
```

<br>

# Pie, donut, and treemap examples
---

#### Pie chart

```
# Make pie chart using chart_data
plt.pie(x=chart_data['Value'], labels=chart_data['Category'] + ": " + chart_data['Value'].astype(str))
plt.show()
```

<br>

### Donut chart
---

```
# Make a donut chart with chart_data
plt.pie(x=chart_data['Value'], labels=chart_data['Category'] + ": " + chart_data['Value'].astype(str))

# Add a circle at the center of the pie chart
my_circle=plt.Circle((0,0), 0.7, color='white')      # Make a circle
p=plt.gcf()                                          # Save the pie chart in variable p 
p.gca().add_artist(my_circle)                        # Add the circle on top of the pie chart


plt.show()
```

<br>

### Treemap
---

```
# Build a treemap using squarify.plot()
squarify.plot(sizes=chart_data['Value']
              , label=chart_data['Category'] + ": " + chart_data['Value'].astype(str)
              , alpha=0.5)

plt.title('Treemap: Size of each category')
plt.axis('off')
plt.show()
```
