## **Python:**


*   Uses matplotlib.pyplot package

<br>

**Vertical bars:**


    plt.bar(x='col1', height='col2', data=df)
    plt.title('Insert title here')
    plt.xlabel('col1 label')
    plt.ylabel('col2 label')
    plt.show()

**Horizontal bars:** 

    plt.barh(y='col2', width='col1', data=df)
    plt.title('Insert title here')
    plt.ylabel('col2 label')
    plt.xlabel('col1 label')
    plt.show()

### Load required packages and data
---

```
# Import required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
# Save Github location paths to a variable
failed_bank_path = 'https://github.com/The-Calculated-Life/python_analysis_for_excel/blob/main/data/failed_banks.xlsx?raw=true'
bx_books_path = 'https://raw.githubusercontent.com/The-Calculated-Life/python_analysis_for_excel/main/data/bx_books.csv'
bx_ratings_path = 'https://raw.githubusercontent.com/The-Calculated-Life/python_analysis_for_excel/main/data/bx_ratings.csv'
bx_users_path = 'https://raw.githubusercontent.com/The-Calculated-Life/python_analysis_for_excel/main/data/bx_users.csv'

# Read excel and CSV files
bank_detail = pd.read_excel(failed_bank_path, sheet_name='detail')
bank_list = pd.read_excel(failed_bank_path, sheet_name='banks')
bank_dividends = pd.read_excel(failed_bank_path, sheet_name='dividends')
bx_books = pd.read_csv(bx_books_path)
bx_ratings = pd.read_csv(bx_ratings_path)
bx_users = pd.read_csv(bx_users_path)

# Load data for chart
data = {'Category':['A', 'B', 'C', 'D', 'E', 'F', 'G'],
        'Value':[12, 8, 10, 6, 7, 8, 10]}

chart_data = pd.DataFrame(data)

# Create dataframe for quick challenge #1
bx_books_count = bx_books[bx_books['year_of_publication'].isin([2000, 2001, 2002, 2003])].groupby('year_of_publication')['isbn'].count()
bx_books_count = pd.DataFrame(bx_books_count).reset_index()
bx_books_count.columns = ['year_of_publication', 'published_count']
bx_books_count['year_of_publication'] = bx_books_count['year_of_publication'].astype(str)
```

<br>

# Examples: Column chart
---

```
# Inspect chart_data
chart_data
```

<br>

```
# change the size of the plot: Applies to all future plots
plt.rcParams['figure.figsize'] = [12,6]
```

<br>

```
# Make column chart using matplotlib
plt.bar(x='Category', height='Value', data=chart_data)
plt.show()
```

<br>

```
# Make column chart using matplotlib: add title
plt.bar(x='Category', height='Value', data=chart_data)
plt.title('Value by Category')
plt.show()
```

<br>

```
# Make column chart using matplotlib: add axis labels
plt.bar(x='Category', height='Value', data=chart_data)
plt.title('Value by Category')
plt.xlabel('Category')
plt.show()
```

<br>

```
# Make column chart using matplotlib: add axis labels
plt.bar(x='Category', height='Value', data=chart_data)
plt.title('Value by Category')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

<br>

```
# Use pandas .plot(kind='bar') method approach instead
chart_data.plot(kind='bar', x='Category', y='Value', title='Category vs Value')
```

<br>

### Horizontal bars 
---

```
# Make a horizonal bar chart using matplotlib
plt.barh(y='Category', width='Value', data=chart_data)
plt.title('Value by Category')
plt.xlabel('Value')
plt.ylabel('Category')
plt.show()
```

<br>

```
# Use pandas .plot(kind='barh') method approach instead
chart_data.plot(kind='barh', x='Category', y='Value', title='Category vs Value')
```
