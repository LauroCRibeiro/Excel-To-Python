### Load required packages and data
---

```
# Import required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
# Save Github location paths to a variable
census_path = 'https://github.com/The-Calculated-Life/python_analysis_for_excel/blob/main/data/census_data.xlsx?raw=true'

# Read excel and CSV files
census = pd.read_excel(census_path)

from vega_datasets import data
cars = data.cars()
cars['Year'] = cars['Year'].dt.year
```

<br> 

### Example: scatterplot
---

```
# View the cars dataframe
cars.head()
```

<br>

```
# change the size of the plot: Applies to all future plots in this notebook
plt.rcParams['figure.figsize'] = [12,6]
```

<br>

```
# Create a scatterplot of "Horsepower" and "Acceleration"
plt.scatter(x='Horsepower', y='Acceleration', data=cars)
plt.show()
```

<br>

```
# Create a scatterplot of "Horsepower" and "Acceleration" (add titles and labels)
plt.scatter(x='Horsepower', y='Acceleration', data=cars)
plt.title('Horsepower vs. Acceleration')
plt.xlabel('Horsepower')
plt.ylabel('Acceleration')
plt.show()
```

<br>

```
# Build the same scatterplot with the pandas approach
cars.plot.scatter(x='Horsepower'
                  , y='Acceleration'
                  , title='Horsepower vs. Acceleration'
                  , xlabel='Horsepower'
                  , ylabel='Acceleration')
```
