## **Python:**



*   Uses matplotlib.pyplot package
*   Can also use pandas approach

<br>

**Matplotlib histogram:**


    plt.hist(x='col1', data=df)
    plt.title('Insert title here')
    plt.xlabel('col1 label')
    plt.ylabel('col2 label')
    plt.show()

<br>

**Pandas approach:**

    df.hist(column='col1')     # One column
    df.plot.hist()             # Every int/float column

### Load required packages and data
---

```
# Import required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
# Save Github location paths to a variable
failed_bank_path = 'https://github.com/The-Calculated-Life/python_analysis_for_excel/blob/main/data/failed_banks.xlsx?raw=true'
bx_ratings_path = 'https://raw.githubusercontent.com/The-Calculated-Life/python_analysis_for_excel/main/data/bx_ratings.csv'

# Read excel and CSV files
bank_detail = pd.read_excel(failed_bank_path, sheet_name='detail')
bx_ratings = pd.read_csv(bx_ratings_path)

# Create filtered versions of original dataframes
bank_detail_small = bank_detail[(bank_detail['ASSETS'] > 0) & (bank_detail['ASSETS'] < 150000)]
bx_ratings_over_zero = bx_ratings[bx_ratings['rating'] > 0]
```

<br>

### Histogram examples
---

```
# View "bank_detail"
bank_detail_small.head()
```

<br>

```
# change the size of the plot: Applies to all future plots in this notebook
plt.rcParams['figure.figsize'] = [12,6]
```

<br>

```
# Make histogram of "bank_detail_small" 
plt.hist(x='ASSETS', data=bank_detail_small)
plt.show()
```

<br>

```
# Make histogram of "bank_detail_small" (add bins arg)
plt.hist(x='ASSETS', data=bank_detail_small, bins=7)
plt.show()
```

<br>

```
# Make histogram of "bank_detail_small" (add edgecolor, linewidth)
plt.hist(x='ASSETS'
        , data=bank_detail_small
        , bins=7
        , edgecolor='black'
        , linewidth=1)
plt.show()
```

<br>

```
# Make histogram of "bank_detail_small" (add title, labels)
plt.hist(x='ASSETS'
        , data=bank_detail_small
        , bins=7
        , edgecolor='black'
        , linewidth=1)

plt.title('Assets Histogram')
plt.xlabel('Assets ($)')
plt.ylabel('# of Banks')
plt.show()
```

<br>

```
# Make histogram of "bank_detail_small" using the Pandas approach
bank_detail_small.hist(column='ASSETS'
                       , bins=7
                       , edgecolor='black'
                       , linewidth=1
                       , grid=False
                       )
```

<br>

```
# Make histogram of all columns on one chart using df.plot.hist()
bank_detail_small.plot.hist(bins=7
                       , edgecolor='black'
                       , linewidth=1
                       , grid=False
                       , alpha=0.5
                       )
```
